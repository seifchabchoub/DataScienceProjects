```python
def getCountry(location) :
    country = []
    
    for x in location : 
        try :
            words = x.split()
            x = words[-1]
        except :
            x =None
        country.append(x)
    return country
```

```python
def firstWord(skill) :
    list=[]
    for i in skill :
        i = i.split()[0]
        list.append(i)
    return list
```

```python
word_vectors = KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin',binary=True)
```

```python
def wordEmbedding(skill):
    skill_ret=[]
    
    for i in skill:
        try :
            i=word_vectors.get_vector(i)
        except :
            i=None
        skill_ret.append(i)
    return skill_ret
```

```python
skill1_w2v = wordEmbedding(dataframe['skill1'])
skill2_w2v = wordEmbedding(dataframe['skill2'])
skill3_w2v = wordEmbedding(dataframe['skill3'])
skill4_w2v = wordEmbedding(dataframe['skill4'])
skill5_w2v = wordEmbedding(dataframe['skill5'])
```

```python
skill1 = pd.DataFrame(list(map(np.ravel, skill1_w2v)))
skill2 = pd.DataFrame(list(map(np.ravel, skill2_w2v)))
skill3 = pd.DataFrame(list(map(np.ravel, skill3_w2v)))
skill4 = pd.DataFrame(list(map(np.ravel, skill4_w2v)))
skill5 = pd.DataFrame(list(map(np.ravel, skill5_w2v)))
```

```python
skill1["skill1"] = skill1.sum(axis=1)
skill2["skill2"] = skill2.sum(axis=1)
skill3["skill3"] = skill3.sum(axis=1)
skill4["skill4"] = skill4.sum(axis=1)
skill5["skill5"] = skill5.sum(axis=1)
```

```python
dataframe_final = pd.concat([skill1["skill1"],skill2["skill2"],skill3["skill3"],skill4["skill4"],skill5["skill5"],Company], axis=1, join='inner')
```

```python
X = dataframe_final.iloc[:,0:-1].values
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters =20 , init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```
